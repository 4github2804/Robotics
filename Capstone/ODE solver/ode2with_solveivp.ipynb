```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

### Solve ODE with Single Solution Component

```python
def odeSSC(t,y):
    return 2*t
print(np.linspace(0,5,100).shape)
sol = solve_ivp(odeSSC,[0,5],[0],t_eval=np.linspace(0,5,100))
```

```python
plt.plot(sol.t,sol.y[0])
```

### Solve Nonstiff Equation

```python
def vdp1(t, y):
    return np.array([y[1], (1 - y[0]**2)*y[1] - y[0]])
```

```python
t = np.linspace(0,20,100)
sol = solve_ivp(vdp1,[0,20],[2,0],t_eval=t)
```

```python
plt.plot(sol.t,sol.y[0])
plt.plot(sol.t,sol.y[1])
```

### Pass Extra Parameters to ODE function

```python
def odefcn(t,y,A=1,B=2):
    return np.array([y[1],A/B*t*y[0]])
```

```python
t = np.linspace(0,5,100)
sol = solve_ivp(odefcn,[0,5],[0,.01],t_eval=t)
```

```python
plt.plot(sol.t,sol.y[0])
plt.plot(sol.t,sol.y[1])
```

### ODE with Time-Dependent Terms

```python
ft = np.linspace(0,5,25)
gt = np.linspace(1,6,25)
f = ft**2-ft-3
g = 3*np.sin(gt-.25)
t0, t1, ic, y0 = 1, 5, 1, 1
t= np.linspace(t0,t1,100)
def myode(t,y,ft,f,gt,g):
    f = np.interp(t,ft,f)
    g = np.interp(t,gt,g)
    dydt = -f*y+g
    return dydt
```

```python
sol=solve_ivp(lambda t,y:myode(t,y,ft,f,gt,g),[t0,t1],
             [y0],t_eval=t)
```

```python
plt.plot(sol.t,sol.y[0])
```

### Capstone

```python
def dyn(t,X,beta):
    delta,alpha,gamma,omega = 1,1,1,1
    x = X[0]; xd = X[1]
    return np.array([xd,gamma*np.cos(omega*t)-alpha*x-delta*xd-beta*x**3])
```

```python
def simode(beta,tend):
    t0, t1 = 0, tend                # start and end
    t = np.linspace(t0, t1, 100)  # the points of evaluation of solution
    y0 = [1, 1]                   # initial value
    sol=solve_ivp(lambda t,X:dyn(t,X,beta),
                 [t0,t1],[1,1],t_eval=t)
    return sol
```

```python
sol = simode(1,50)
plt.figure(figsize=(20,5))
plt.subplot(1,2,1); plt.plot(sol.t,sol.y[0]);plt.plot(sol.t,sol.y[1])
plt.subplot(1,2,2); plt.plot(sol.y[0],sol.y[1])
```
