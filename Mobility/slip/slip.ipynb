```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.integrate as inte
from numpy import matmul as mm
from mpl_toolkits import mplot3d
```

```python
xd0 = .5 #initial speed along x axis (x dot)
tdang = 0 #angle (randian)
def loadSettings(xd0,tdang):
    p = {} # 
    p['mb'],p['omega'],p['rho']=1,50,.12
    p['g'],p['b']=9.81,20
    p['tdang'],p['xd0']=tdang,xd0
    p['ttd1'] = 0
    
    ic = np.array([0,
          p['rho'],p['tdang'],-.8,.2,
          0,0,p['xd0'],0])
    tmax = 2
    return p,ic,tmax
```

```python
def unpackX(X):
    if X.size>9:
        r1,th1 = X[:,1],X[:,2]
        x,z = X[:,3],X[:,4]
        dr1,dth1 = X[:,5],X[:,6]
        dx,dz = X[:,7],X[:,8]
    else:
        r1,th1,x,z = X[1],X[2],X[3],X[4]
        dr1,dth1,dx,dz=X[5],X[6],X[7],X[8]
    return r1,th1,x,z,dr1,dth1,dx,dz
```

```python
def dynamics(t,X,p):
    Xd,u = X,0
    Xd[0] = 0
    Xd[1:5] = X[5:]
    r1,th1,x,z,dr1,dth1,dx,dz = unpackX(X)
    mb,g = p['mb'],p['g']
    ang1 = math.atan2(-(r1-p['rho'])*p['omega'],-dr1)
    if X[0] == 0:
        Xd[6],Xd[5] = 0,0
    else:
        print('bye')
        u = -p['omega']**2*(r1-p['rho'])
        
    if X[0] == 0:
        Xd[7],Xd[8] = 0,-p['g']
    elif X[0] == 1:
        rhs = [dth1**2*r1+mb**(-1)*u-g*np.cos(th1),
               r1**(-1)*(-2*dr1*dth1+g*np.sin(th1)),
               -mb**(-1)*u*np.sin(th1),
               -g+mb**(-1)*u*np.cos(th1),u*np.sin(th1)]
        Xd[5:9] = rhs[:4]
    return Xd
```

```python
def runSim(tmax,ic,p):
    ta,ya,tLatest = [],[],0
    while tLatest < tmax:
        tspan = np.array([tLatest,tmax])
        print(tspan[0])
        tChart = np.linspace(tspan[0],.01,100)
        yChart = np.zeros((len(tChart),len(ic)))
        yChart[0,:] = ic
        r = inte.ode(lambda t,X: dynamics(t,X,p))\
            .set_integrator("dopri5",max_step=.01)
        r.set_initial_value(ic,tChart[0])
        for i in range(1,tChart.size):
            yChart[i,:] = r.integrate(tChart[i]).flatten()
            if not r.successful():
                raise RuntimeError("Could not integrate")
        print(yChart)
        break
    
    return ta,ya
```

```python
def slip(xd0,tdang):
    p,ic,tmax = loadSettings(xd0,tdang)
    t,X = runSim(tmax,ic,p)
    return t,X
```
